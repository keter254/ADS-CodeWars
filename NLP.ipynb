```python
# NLP  workd with unstructured data
#Get some text data from twitter/facebook/blogs,news
import requests
from bs4 import BeautifulSoup
URL="https://www.standardmedia.co.ke/sports/football/article/2001462653/samuel-etoo-sorry-after-attacking-a-man-following-brazils-4-1-win-over-south-korea-at-world-cup"
page=requests.get(URL)
soup=BeautifulSoup(page.content,"html.parser")
div=soup.findAll("div",class_="story")
soup=BeautifulSoup(str(div))
text =soup.get_text()
#print(text)
```

```python
import nltk
nltk.download("punkt")
```

```python
#convert text to sentences or works
from nltk.tokenize import  sent_tokenize,word_tokenize
tokenized_text =word_tokenize(text)
print(tokenized_text)
```

```python
nltk.download("stopwords")
from nltk.corpus import stopwords
print(stopwords.fileids())
#Get english stopwords
mystopwords = set(stopwords.words('english'))
print(mystopwords)
```

```python
#justpasteit/bi8nu
filtered_words =[]
for word in tokenized_text:
       if word.lower() not in mystopwords:
               filtered_words.append(word)

print("Old Words ", tokenized_text)
print("New Words ", filtered_words)


```

```python
#justpasteit/biph5
clean_words = [word for word in filtered_words if word.isalpha()]
print(clean_words)
```

```python
from nltk.probability import FreqDist
frequency =FreqDist(clean_words)
frequency.most_common(10)
```

```python
#Justpasteit/cq844
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
#Function to Create Wordcloud
def create_wordcloud(text):
        stopwords = set(STOPWORDS)
        wc = WordCloud(background_color="white",
        max_words=3000,
        stopwords=stopwords,repeat=True)
        wc.generate(str(text))
        wc.to_file("wc.png")
        print("Word Cloud Saved Successfully")
        path="wc.png"
        display(Image.open(path))
```

```python
create_wordcloud(frequency.most_common(10))
```
